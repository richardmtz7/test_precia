```python
import pandas as pd

def clean_data(df):
    # Filter rows based on column: 'Pclass'
    df = df[df['Pclass'] == 1]
    # Sort by column: 'Name' (ascending)
    df = df.sort_values(['Name'])
    # Drop columns: 'PassengerId', 'Survived' and 8 other columns
    df = df.drop(columns=['PassengerId', 'Survived', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'])
    return df

# Loaded variable 'df' from URI: c:\Users\liset\OneDrive\Escritorio\Doc Richard\prueba_precia\prueba_tecnica\data\titanic.csv
df = pd.read_csv(r'c:\\Users\\liset\\OneDrive\\Escritorio\\Doc Richard\\prueba_precia\\prueba_tecnica\\data\\titanic.csv')

df_clean = clean_data(df.copy())
df_clean.head(10)
df_clean.head(-10)
```

```python
survivors = df["Survived"].value_counts(normalize=True) * 100
# Donde 0 es el porcentaje que murió y 1 el que sobrevivió
survivors.round()
```

```python
survivors_by_class = df.groupby("Pclass")["Survived"].mean()*100
# Se muestran los porcentajes por clase
survivors_by_class.round()
```

```python
drop_unknown_values = df.dropna(subset=['Age'])
drop_unknown_values
```

```python
women = df[df["Sex"] == "female"]

middle_ages_by_class = women.groupby("Pclass")["Age"].mean()
# Se muestra la edad media de las mujeres por clases
middle_ages_by_class.round()
```

```python
# Para un mejor análisis, primero se eliminar los datos faltantes
df.dropna(subset=['Age'])

#Se aplica la condición, suponiendo que se considera mayor de edad cuando tienen 18 anios en adelante
df["Era_menor_De_Edad"] = df["Age"] >= 18
df["Era_menor_De_Edad"]
```

```python
df['Edad_Categoria'] = pd.cut(df['Age'], bins=[0, 17, float('inf')], labels=['Menor', 'Mayor'])

# Calcular el porcentaje de menores y mayores de edad que sobrevivieron en cada clase
porcentaje_sobrevivientes_por_edad_clase = df.groupby(['Pclass', 'Edad_Categoria'])['Survived'].mean() * 100
porcentaje_sobrevivientes_por_edad_clase.round()
```
